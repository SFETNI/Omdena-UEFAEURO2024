```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
league_table = pd.read_csv("/Users/tsovinarbabakhanyan/Desktop/TunisiaLocalChapter_UEFAEURO2024/task2-data-analysis/Cleaned_Datsets/League_table2021.csv")
```

```python
league_table.head()
```

```python
plt.plot(figsize=(16,6))
sns.pairplot(data=league_table,diag_kind='kde')
```

```python
league_table = league_table.dropna()
```

```python
league_table
```

```python
x = league_table.drop(['Matches Played'],axis=1)
y = league_table['Matches Played']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
randclass = RandomForestClassifier(n_estimators = 20, random_state = 0)
randclass.fit(X_train,Y_train)
y_pred = randclass.predict(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression(max_iter=100)
log.fit(X_train,Y_train)
```

```python
from sklearn.svm import LinearSVC
svc = LinearSVC()
svc.fit(X_train,Y_train)
```

```python
from sklearn import tree
treeclass = tree.DecisionTreeClassifier()
treeclass = treeclass.fit(X_train, Y_train)
```

```python
from sklearn.metrics import classification_report
def report(model):
    preds = model.predict(X_test)
    print(classification_report(Y_test,preds))
```

```python
print("Logistic Regression")
report(log)
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(Y_test, y_pred)
plt.show()
```

```python
print("SVC")
report(svc)
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(Y_test, y_pred)
plt.show()
```

```python
print("Random Forest")
report(randclass)
ConfusionMatrixDisplay.from_predictions(Y_test, y_pred)
plt.show()
```

```python
print("Decition tree")
report(treeclass)
ConfusionMatrixDisplay.from_predictions(Y_test, y_pred)
plt.show()
```

```python
# plot tree
plt.figure(figsize=(50,50))  # set plot size (denoted in inches)
tree.plot_tree(treeclass, fontsize=20)
plt.show()
```
