```python
import pandas as pd

# Read the CSV file
data = pd.read_csv('johnpaul-euro_2021_stats.csv')

# Display the first few rows of the data
print(data.head())

# Display basic information about the data
print(data.info())

# Display summary statistics of the data
print(data.describe())
```

```python
missing_values = data.isnull().sum()
print("Missing values:\n", missing_values)

# Drop rows with missing values (if any)
data = data.dropna()
```

```python
# Remove duplicate rows
data = data.drop_duplicates()
```

```python
print(data.head())
```

```python
data.to_csv('cleaned_euro_2021_stats.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the cleaned CSV file
data = pd.read_csv('cleaned_euro_2021_stats.csv')

# Display basic information about the data
print(data.info())

# Display summary statistics of the data
print(data.describe())

# Display the first few rows of the data
print(data.head())
```

```python
# Scatter plot of two numerical columns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='xG', y='GF', data=data)
plt.title('Scatter Plot of xG vs GF')
plt.show()

# Pair plot for a subset of columns
sns.pairplot(data[['xG', 'GF', 'xGA', 'GA', 'Pts']])
plt.show()
```

```python
# Calculate the correlation matrix
corr_matrix = data.corr()

# Display the correlation matrix
print(corr_matrix)

# Plot a heatmap of the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Top 5 teams by points
top_teams_by_points = data[['Squad', 'Pts']].sort_values(by='Pts', ascending=False).head(5)
print("Top 5 teams by points:\n", top_teams_by_points)

# Top 5 teams by goals scored
top_teams_by_goals = data[['Squad', 'GF']].sort_values(by='GF', ascending=False).head(5)
print("Top 5 teams by goals scored:\n", top_teams_by_goals)
```
