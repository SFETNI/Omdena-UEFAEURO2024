```python
#Here is one of the issentiols are the XGBOost part for the predicting arts as well as for testing and the residuals
import xgboost as xg
from sklearn.metrics import mean_squared_error as MSE
from sklearn.model_selection import train_test_split
```

```python
X, y = euro_team_numerical.iloc[:, :-1], euro_team_numerical.iloc[:, -1]
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y,
                      test_size = 0.3, random_state = 12)
```

```python
xgb_r = xg.XGBRegressor(objective ='reg:linear',
                  n_estimators = 10, seed = 12)
```

```python
# Fitting the model
xgb_r.fit(train_X, train_y)

# Predict the model
pred = xgb_r.predict(test_X)

# RMSE Computation
rmse = np.sqrt(MSE(test_y, pred))
print("RMSE : % f" %(rmse))
```

```python
residuals = test_y - pred
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(residuals,pred)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')  # Line for perfect predictions
plt.title('Residuals VS Pred')
plt.xlabel('Pred')
plt.ylabel('Residuals')
plt.show()
```
