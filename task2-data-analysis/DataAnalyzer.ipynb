```python
data.drop(['Date', 'HomeTeam', 'AwayTeam', 'Tournament', 'CityVenue', 'CountryVenue'], axis=1, inplace=True)
plt.figure(figsize=(20,10))
heatmapChart=data.corr()
sns.heatmap(heatmapChart, cmap="BrBG", annot=True)
heatmapChart
```

```python
data = pd.read_csv('Data/clean-data/clean-data.csv')
fig, ax = plt.subplots(figsize=(20,10))
ax.scatter(data['FTHG'], data['FTAG'])
ax.set_xlabel('Full Time Home Team Goals')
ax.set_ylabel('Full Time Away Team Goals')
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
data = pd.DataFrame({
    x_label: grp['AwayTeam'].value_counts()
    for x_label, grp in data.groupby('HomeTeam')
})
sns.heatmap(data, cmap='viridis')
plt.xlabel('HomeTeam')
_ = plt.ylabel('AwayTeam')
```

```python
data.groupby('AwayTeam').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```
