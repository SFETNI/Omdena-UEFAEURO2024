Missing Values of coundition_pitch column are replaced by the most frequent entry of the column

```python
condition_pitch_list = df['condition_pitch'].tolist()
```

```python
from collections import Counter
import numpy as np
data_array = np.array(condition_pitch_list, dtype=object)
data_array
```

```python
most_frequent = Counter(data_array[~pd.isna(data_array)]).most_common(1)[0][0]
most_frequent
```

```python
df['condition_pitch'] = df['condition_pitch'].fillna(most_frequent)
```

```python
# Check for null values in each column
null_values = df.isnull().sum()

# Print the results
print(null_values)
```

```python
condition_weather_list = df['condition_weather'].tolist()
condition_weather_list
```

```python
# Convert the list to a numpy array for convenience
data_array = np.array(condition_weather_list, dtype=object)

# Find the most frequent non-nan value
most_frequent = Counter(data_array[~pd.isna(data_array)]).most_common(1)[0][0]
most_frequent
```

```python
df['condition_weather'] = df['condition_weather'].fillna(most_frequent)
```

```python
# Check for null values in each column
null_values = df.isnull().sum()

# Print the results
print(null_values)
```

**Heatmaps to visualize the frequency distributions of match conditions like pitch and weather across different categories such as matchday, condition weather, and group name. These visualizations help in understanding how match conditions vary with different groupings.**

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd

fig, axes = plt.subplots(2, 2, figsize=(16, 16))

# First heatmap
df_2dhist_1 = pd.DataFrame({
    x_label: grp['condition_pitch'].value_counts()
    for x_label, grp in df.groupby('matchday_name')
})
sns.heatmap(df_2dhist_1, cmap='viridis', ax=axes[0, 0])
axes[0, 0].set_xlabel('matchday_name')
axes[0, 0].set_ylabel('condition_pitch')
axes[0, 0].set_title('Heatmap 1: matchday_name vs condition_pitch')

# Second heatmap
df_2dhist_2 = pd.DataFrame({
    x_label: grp['condition_weather'].value_counts()
    for x_label, grp in df.groupby('matchday_name')
})
sns.heatmap(df_2dhist_2, cmap='viridis', ax=axes[0, 1])
axes[0, 1].set_xlabel('matchday_name')
axes[0, 1].set_ylabel('condition_weather')
axes[0, 1].set_title('Heatmap 2: matchday_name vs condition_weather')

# Third heatmap
df_2dhist_3 = pd.DataFrame({
    x_label: grp['condition_pitch'].value_counts()
    for x_label, grp in df.groupby('condition_weather')
})
sns.heatmap(df_2dhist_3, cmap='viridis', ax=axes[1, 0])
axes[1, 0].set_xlabel('condition_weather')
axes[1, 0].set_ylabel('condition_pitch')
axes[1, 0].set_title('Heatmap 3: condition_weather vs condition_pitch')

# Fourth heatmap
df_2dhist_4 = pd.DataFrame({
    x_label: grp['condition_weather'].value_counts()
    for x_label, grp in df.groupby('group_name')
})
sns.heatmap(df_2dhist_4, cmap='viridis', ax=axes[1, 1])
axes[1, 1].set_xlabel('group_name')
axes[1, 1].set_ylabel('condition_weather')
axes[1, 1].set_title('Heatmap 4: group_name vs condition_weather')

# Adjust layout
plt.tight_layout()
plt.show()
```

**Violin plots to visualize the distribution of home and away scores against various categorical variables such as stadium, matchday, and group. These plots provide insights into the spread and density of scores across different conditions.**

```python
import matplotlib.pyplot as plt
import seaborn as sns


rows = 3
cols = 2

figsize = (12, 2 * len(df['group_name'].unique()))
plt.figure(figsize=figsize)

plot_titles = [
    "Home Score vs Stadium",
    "Away Score vs Stadium",
    "Away Score vs Matchday",
    "Home Score vs Matchday",
    "Away Score vs Group",
    "Home Score vs Group"
]

# Loop through each position in the grid and create a violin plot
count = 1
for i in range(rows):
    for j in range(cols):
        if count > len(plot_titles):
            break

        if count <= 2:
            x_var = (['home_score', 'away_score'][count - 1])
            y_var = 'stadium_name'
            palette = 'Dark2'
        elif count <= 4:
            x_var = 'matchday_name'
            y_var = (['away_score', 'home_score'][count - 3])
            palette = 'viridis'
        else:
            x_var = 'group_name'
            y_var = (['away_score', 'home_score'][count - 5])
            palette = 'inferno'

        # Create the subplot at the current position
        ax = plt.subplot(rows, cols, count)

        # Create the violin plot using seaborn
        sns.violinplot(data=df, x=x_var, y=y_var, inner='quartile', palette=palette, ax=ax)
        sns.despine(top=True, right=True)

        # Add title for each subplot
        ax.set_title(plot_titles[count - 1], fontsize=10)
        ax.set_xlabel(x_var, fontsize=8)  # Set x-axis label
        ax.set_ylabel(y_var, fontsize=8)  # Set y-axis label
        count += 1

# Adjust layout to prevent overlapping elements
plt.tight_layout()

# Display the plot
plt.show()
```

**Bar plots to analyze match attendance based on various factors such as home team score, winner, away team score, stadium name, and match round. It also visualizes the total stadium capacity by stadium name.**

```python
import matplotlib.pyplot as plt
import seaborn as sns


fig, axs = plt.subplots(3, 2, figsize=(16, 14))  # Create a 3x2 grid of subplots

# Define color palettes
inferno_palette = sns.color_palette('inferno', n_colors=10)
plasma_palette = sns.color_palette('plasma', n_colors=10)
viridis_palette = sns.color_palette('viridis', n_colors=10)
dark2_palette = sns.color_palette('Dark2', n_colors=10)

# Plot 1: Match Attendance vs Home Team Score
df.groupby('home_score')['match_attendance'].sum().sort_values(ascending=False).plot(
    kind='bar', color=inferno_palette, ax=axs[0, 0]
)
axs[0, 0].spines[['top', 'right']].set_visible(False)
axs[0, 0].set_xlabel('Home Team Score')
axs[0, 0].set_ylabel('Match Attendance')
axs[0, 0].set_title('Match Attendance vs Home Team Score')

# Plot 2: Match Attendance vs Winner
df.groupby('winner')['match_attendance'].sum().sort_values(ascending=False).plot(
    kind='bar', color=inferno_palette, ax=axs[0, 1]
)
axs[0, 1].spines[['top', 'right']].set_visible(False)
axs[0, 1].set_xlabel('Winner')
axs[0, 1].set_ylabel('Match Attendance')
axs[0, 1].set_title('Match Attendance vs Winner')

# Plot 3: Match Attendance vs Away Team Score
df.groupby('away_score')['match_attendance'].sum().sort_values(ascending=False).plot(
    kind='bar', color=plasma_palette, ax=axs[1, 0]
)
axs[1, 0].spines[['top', 'right']].set_visible(False)
axs[1, 0].set_xlabel('Away Team Score')
axs[1, 0].set_ylabel('Match Attendance')
axs[1, 0].set_title('Match Attendance vs Away Team Score')

# Plot 4: Total Stadium Capacity vs Stadium Name
df.groupby('stadium_name')['stadium_capacity'].sum().sort_values(ascending=False).plot(
    kind='bar', color=viridis_palette, ax=axs[1, 1]
)
axs[1, 1].spines[['top', 'right']].set_visible(False)
axs[1, 1].set_xlabel('Stadium Name')
axs[1, 1].set_ylabel('Total Capacity')
axs[1, 1].set_title('Total Stadium Capacity vs Stadium Name')

# Plot 5: Match Attendance vs Stadium Name
df.groupby('stadium_name')['match_attendance'].sum().sort_values(ascending=False).plot(
    kind='bar', color=dark2_palette, ax=axs[2, 0]
)
axs[2, 0].spines[['top', 'right']].set_visible(False)
axs[2, 0].set_xlabel('Stadium Name')
axs[2, 0].set_ylabel('Match Attendance')
axs[2, 0].set_title('Match Attendance vs Stadium Name')

# Plot 6: Match Attendance vs Round
df.groupby('round')['match_attendance'].sum().sort_values(ascending=False).plot(
    kind='bar', color=inferno_palette, ax=axs[2, 1]
)
axs[2, 1].spines[['top', 'right']].set_visible(False)  # Remove top and right spines
axs[2, 1].set_xlabel('Round')
axs[2, 1].set_ylabel('Match Attendance')
axs[2, 1].set_title('Match Attendance vs Round')

plt.tight_layout()

# Display the plot
plt.show()
```

**Swarm Plots to visualize the distributions and relationships between categorical variables in soccer matches. The heatmaps show how pitch conditions and weather vary across different match days, and how weather conditions relate to both pitch conditions and group names, providing insights into potential correlations or patterns in the data.**

```python

fig, axes = plt.subplots(2, 2, figsize=(16, 16))

# Define the parameters for each swarm plot
parameters = [
    {'x': 'home_score', 'y': 'away_score', 'hue': 'winner', 'palette': 'viridis'},
    {'x': 'condition_temperature', 'y': 'condition_humidity', 'hue': 'stadium_name', 'palette': 'plasma'},
    {'x': 'matchday_name', 'y': 'winner', 'hue': 'group_name', 'palette': 'inferno'},
    {'x': 'round', 'y': 'winner', 'hue': 'condition_weather', 'palette': 'Dark2'},
]

for i, parameter in enumerate(parameters):
    ax = axes.flat[i]
    sns.swarmplot(data=df, x=parameter['x'], y=parameter['y'], hue=parameter['hue'], palette=parameter['palette'], ax=ax)
    ax.set_title(f'Swarm Plot of {parameter["y"]} vs. {parameter["x"]}')
    ax.set_xlabel(parameter['x'])
    ax.set_ylabel(parameter['y'])

plt.tight_layout()
plt.show()
```
