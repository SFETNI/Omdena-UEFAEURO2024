```python
euro_team_numerical = euro_team[['Age','Height','Goals','MarketValue','Caps']]
```

```python
sns.heatmap(euro_team_numerical.corr(),annot=True, fmt='.2f', cmap='Pastel2', linewidths=2)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.clustermap(data=euro_team_numerical.corr(),z_score=0, cmap="vlag", center=0,annot=True)
```

```python
#lets do the clustering exploratory data analasys
from sklearn.cluster import KMeans
clusters = 12

kmeans = KMeans(n_clusters = clusters)
kmeans.fit(euro_team_numerical)

print(kmeans.labels_)
```

```python
#some PCA as well
from sklearn.decomposition import PCA

pca = PCA(5)
pca.fit(euro_team_numerical)

pca_data = pd.DataFrame(pca.transform(euro_team_numerical))

print(pca_data.head())
```
