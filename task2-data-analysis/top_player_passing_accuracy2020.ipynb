```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('euro2020.csv')
data.sample(50)
```

```python
data.shape
```

```python
data.info()
```

```python
data.fillna(0)
```

```python
passing_accuracy_top_player = data['Country']
plt.figure(figsize=(20,10))
plt.bar(passing_accuracy_top_player,passing_accuracy_top_player.index, color='green')
plt.title('passing accuracy of top player from each country 2020')
plt.xlabel('Country')
plt.ylabel('Passing accuracy')
plt.show()
```
