## Data cleaning

**Check the data types of the columns for the dataset**

Lets see the detected type for each of the columns and compare it with how the actual data looks. Also, we can see how many nulls and not nulls values exist

```python
data.info()
```

There are many columns will null values, among them: home_penalty, away_penalty, winner, group_name, condition_humidity, condition_pitch, condition_temperature, condition_weather, condition_wind_speed, goals, penalties_missed, penalties, red_cards, stadium_name_sponsor. We will treat them depending on the type of the column: numeric or string

**Filling missing values**

```python
# Penalties number are 0 if not present
data['home_penalty'].fillna(0, inplace=True)
data['away_penalty'].fillna(0, inplace=True)
# Winner will be empty if there is a draw. We will use N/A in that case
data['winner'].fillna('N/A', inplace=True)
# No group name. N/A will be used
data['group_name'].fillna('N/A', inplace=True)
# We will use the mode for the humidity
data['condition_humidity'] = data['condition_humidity'].fillna(data['condition_humidity'].mode()[0])
# Condition pitch is a object value. We will use N/A
data['condition_pitch'].fillna('N/A', inplace=True)
# condition_temperature, we will use the mode
data['condition_temperature'] = data['condition_temperature'].fillna(data['condition_temperature'].mode()[0])
# condition_weather is a string, will use N/A
data['condition_weather'].fillna('N/A', inplace=True)
# condition_wind_speed is int. Will use the mode
data['condition_wind_speed'] = data['condition_wind_speed'].fillna(data['condition_wind_speed'].mode()[0])
# goals is a json string with goals metadata. Will use empty array [] when not present
data['goals'].fillna('[]', inplace=True)
# penalties_missed is a json string with penalties metadata. Will use empty array [] when not present
data['penalties_missed'].fillna('[]', inplace=True)
# penalties is a json string with penalties metadata. Will use empty array [] when not present
data['penalties'].fillna('[]', inplace=True)
# red_cards is a json string with red cards metadata. Will use empty array [] when not present
data['red_cards'].fillna('[]', inplace=True)
# stadium_name_sponsor 
data['stadium_name_sponsor'].fillna('N/A', inplace=True)
```

Checking the result of missing values treatment

```python
data.isnull().sum()
```

**Changing data types**

We will convert some columns to a different type that makes more sense. For example, penalties should be int type, not float

```python
# All of these are better as int type, since there aren't fractional data
data = data.astype({"home_penalty": int, "away_penalty": int, "condition_humidity": int, 
                    "condition_temperature": int, "condition_wind_speed": int  })
```

### Goals by country

```python
away_goals_country = data.groupby('away_team')['away_score_total'].sum().reset_index().sort_values(by='away_score_total', ascending=False)
#away_goals_country.sample(5)
home_goals_country = data.groupby('home_team')['home_score_total'].sum().reset_index().sort_values(by='home_score_total', ascending=False)
#home_goals_country.head(10)

total_goals_country = away_goals_country.merge(home_goals_country, how='inner', left_on='away_team', right_on='home_team')
total_goals_country['total_goals'] = total_goals_country['away_score_total']  + total_goals_country['home_score_total'] 
total_goals_country = total_goals_country.rename(columns={'away_team': 'team'})



```

Total goals per country

```python
plt.figure(figsize=(15,10))
goals = sns.barplot(
    data= total_goals_country,
    x= 'total_goals',
    y= 'team')
```
