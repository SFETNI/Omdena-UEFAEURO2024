The missing values of the condition_humidity column is replaced by the mean of the column


```python

condition_humidity_list = df['condition_humidity'].tolist()
```

```python
import numpy as np

# Convert the list to a numpy array for convenience
data_array = np.array(condition_humidity_list)

# Calculate the mean of the available data, ignoring NaNs
mean_value = np.nanmean(data_array)
rounded_mean = round(mean_value)

# Replace NaNs with the mean value
data_array[np.isnan(data_array)] = rounded_mean

# Convert back to a list
result_list = data_array.tolist()
```

```python
# send this list to the condition_humidity column

df['condition_humidity'] = result_list
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have a dataframe 'df' with relevant columns

# Define number of rows and columns for the grid
rows = 3
cols = 2

# Define figure size based on number of groups (adjust if needed)
figsize = (12, 1.6 * len(df['group_name'].unique()))
plt.figure(figsize=figsize)

# List of plot titles (modify as needed)
plot_titles = [
    "Home Score vs Stadium",
    "Away Score vs Stadium",
    "Away Score vs Matchday",
    "Home Score vs Matchday",
    "Away Score vs Group",
    "Home Score vs Group"
]

# Loop through each position in the grid and create a violin plot
count = 1
for i in range(rows):
    for j in range(cols):
        if count > len(plot_titles):
            # No more plots to create, break out
            break

        # Select variables based on plot position
        if count <= 2:
            x_var = (['home_score', 'away_score'][count - 1])
            y_var = 'stadium_name'
            palette = 'Dark2'
        elif count <= 4:
            x_var = 'matchday_name'
            y_var = (['away_score', 'home_score'][count - 3])
            palette = 'viridis'
        else:
            x_var = 'group_name'
            y_var = (['away_score', 'home_score'][count - 5])
            palette = 'inferno'

        # Create the subplot at the current position
        ax = plt.subplot(rows, cols, count)

        # Create the violin plot using seaborn
        sns.violinplot(data=df, x=x_var, y=y_var, inner='quartile', palette=palette, ax=ax)
        sns.despine(top=True, right=True)

        # Add title for each subplot
        ax.set_title(plot_titles[count - 1], fontsize=10)
        ax.set_xlabel(x_var, fontsize=8)  # Set x-axis label
        ax.set_ylabel(y_var, fontsize=8)  # Set y-axis label
        count += 1

# Adjust layout to prevent overlapping elements
plt.tight_layout()

# Display the plot
plt.show()
```

```python

# Assuming you have a dataframe 'df' with relevant columns

fig, axs = plt.subplots(2, 2, figsize=(12, 8))  # Create a 2x2 grid of subplots

# Plot 1: Match Attendance vs Home Team Score
df.groupby('home_score')['match_attendance'].sum().sort_values(ascending=False).plot(
    kind='bar', color=sns.palettes.mpl_palette('inferno'), ax=axs[0, 0]
)
axs[0, 0].spines[['top', 'right']].set_visible(False)  # Remove top and right spines
axs[0, 0].set_xlabel('Home Team Score')
axs[0, 0].set_ylabel('Match Attendance')
axs[0, 0].set_title('Match Attendance vs Home Team Score')

# Plot 2: Match Attendance vs Winner
df.groupby('winner')['match_attendance'].sum().sort_values(ascending=False).plot(
    kind='bar', color=sns.palettes.mpl_palette('inferno'), ax=axs[0, 1]
)
axs[0, 1].spines[['top', 'right']].set_visible(False)  # Remove top and right spines
axs[0, 1].set_xlabel('Winner')
axs[0, 1].set_ylabel('Match Attendance')
axs[0, 1].set_title('Match Attendance vs Winner')

# Plot 3: Match Attendance vs Away Team Score
df.groupby('away_score')['match_attendance'].sum().sort_values(ascending=False).plot(
    kind='bar', color=sns.palettes.mpl_palette('plasma'), ax=axs[1, 0]
)
axs[1, 0].spines[['top', 'right']].set_visible(False)  # Remove top and right spines
axs[1, 0].set_xlabel('Away Team Score')
axs[1, 0].set_ylabel('Match Attendance')
axs[1, 0].set_title('Match Attendance vs Away Team Score')

# Plot 4: Total Stadium Capacity vs Stadium Name
df.groupby('stadium_name')['stadium_capacity'].sum().sort_values(ascending=False).plot(
    kind='bar', color=sns.palettes.mpl_palette('viridis'), ax=axs[1, 1]
)
axs[1, 1].spines[['top', 'right']].set_visible(False)  # Remove top and right spines
axs[1, 1].set_xlabel('Stadium Name')
axs[1, 1].set_ylabel('Total Capacity')
axs[1, 1].set_title('Total Stadium Capacity vs Stadium Name')

# Adjust layout to prevent overlapping elements (optional)
plt.tight_layout()

# Display the plot
plt.show()
```
