```python
#convert the entire row 17 to NaN

import numpy as np
df.iloc[16] = np.nan
df.head(17)
```

```python
# mean of attendance_avg
# mean of red_cards
# replace the missing value with it

attendance_avg_mean = df["attendance_avg"].mean()
df["attendance_avg"] = df["attendance_avg"].fillna(attendance_avg_mean)
df["matches"] = df["matches"].fillna(51)
df["year"] = df["year"].fillna(2024)
red_cards_avg_mean = df["red_cards"].mean()
red_cards_avg_mean_rounded=round(red_cards_avg_mean)
df["red_cards"] = df["red_cards"].fillna(red_cards_avg_mean_rounded)
df.head(17)
```

```python

# Calculate the average annual growth rate of attendance
attendance_non_null = df.dropna(subset=['attendance'])
years = attendance_non_null['year'].values
attendances = attendance_non_null['attendance'].values

# Calculate the average growth rate
growth_rates = np.diff(attendances) / attendances[:-1]
average_growth_rate = np.mean(growth_rates)

# Estimate the attendance for the missing year
last_known_attendance = attendances[-1]
predicted_attendance = last_known_attendance * (1 + average_growth_rate)

# Filling the missing value
df.loc[df['attendance'].isnull(), 'attendance'] = predicted_attendance

# Displaying the DataFrame
pd.set_option('display.float_format', '{:.2f}'.format)
print("DataFrame after using trend to estimate attendance:")
df.head(17)
```

```python

# Calculate the average annual growth rate of goals
goals_non_null = df.dropna(subset=['goals'])
years_goals = goals_non_null['year'].values
goals = goals_non_null['goals'].values

# Calculate the average growth rate
growth_rates_goals = np.diff(goals) / goals[:-1]
average_growth_rate_goals = np.mean(growth_rates_goals)

# Estimate the goals for the missing year
last_known_goals = goals[-1]
predicted_goals = last_known_goals * (1 + average_growth_rate_goals)

# Filling the missing value
df.loc[df['goals'].isnull(), 'goals'] = predicted_goals

# Displaying the DataFrame with readable numbers
pd.set_option('display.float_format', '{:.2f}'.format)
print("DataFrame after using trend to estimate goals:")
df.head(17)
```

```python
import seaborn as sns
sns.pairplot(df)
```

```python
# Create a figure with 2 rows and 2 columns of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Plot each graph on its respective subplot
df.groupby('result').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'), ax=axes[0, 0])
df.groupby('matches').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'), ax=axes[0, 1])
df.groupby('red_cards').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'), ax=axes[1, 0])
df.groupby('winner').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'), ax=axes[1, 1])

# Hide the top and right spines for all subplots
for ax in axes.flat:
    ax.spines[['top', 'right']].set_visible(False)

# Set the x-axis label for all subplots
for ax in axes.flat:
    ax.set_xlabel('frequency')

# Adjust the spacing between subplots
plt.tight_layout()
```

```python
plt.hist(df['goals'], bins=20, edgecolor='white')
plt.title('Histogram of Goals')
plt.xlabel('Goals')
plt.ylabel('Frequency')
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Pie chart for Goals
axs[0].pie(df['goals'], labels=df['year'], autopct='%1.1f%%', startangle=140)
axs[0].set_title('Goals Distribution over the Years')

# Pie chart for Attendance
axs[1].pie(df['attendance'].fillna(0), labels=df['year'], autopct='%1.1f%%', startangle=140)
axs[1].set_title('Attendance Distribution over the Years')

# Pie chart for Average Attendance
axs[2].pie(df['attendance_avg'].fillna(0), labels=df['year'], autopct='%1.1f%%', startangle=140)
axs[2].set_title('Average Attendance Distribution over the Years')

plt.tight_layout()
plt.show()
```

```python

fig, axs = plt.subplots(2, 2, figsize=(16, 12))

# Regression Plot of Goals vs Attendance
sns.regplot(x='attendance', y='goals', data=df, ax=axs[0, 0])
axs[0, 0].set_title('Regression Plot of Goals vs Attendance')
axs[0, 0].set_xlabel('Attendance')
axs[0, 0].set_ylabel('Goals')

# Regression Plot of Matches vs Attendance
sns.regplot(x='attendance', y='matches', data=df, ax=axs[0, 1])
axs[0, 1].set_title('Regression Plot of Matches vs Attendance')
axs[0, 1].set_xlabel('Attendance')
axs[0, 1].set_ylabel('Matches')

# Regression Plot of Average Attendance vs Total Attendance
sns.regplot(x='attendance', y='attendance_avg', data=df, ax=axs[1, 0])
axs[1, 0].set_title('Regression Plot of Average Attendance vs Total Attendance')
axs[1, 0].set_xlabel('Attendance')
axs[1, 0].set_ylabel('Average Attendance')

# Regression Plot of Red Cards vs Attendance
sns.regplot(x='attendance', y='red_cards', data=df, ax=axs[1, 1])
axs[1, 1].set_title('Regression Plot of Red Cards vs Attendance')
axs[1, 1].set_xlabel('Attendance')
axs[1, 1].set_ylabel('Red Cards')

plt.tight_layout()
plt.show()
```
